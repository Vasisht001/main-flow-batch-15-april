```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv("C:\\Users\\lenov\\Downloads\\householdtask3.csv")
```

```python
display(data.head(20))
```

```python
#scatter plot with your against method
plt.scatter(data["year"], data["own"])

#adding title to the plot
plt.title("Scatter PLot")

#setting the x and y lables
plt.xlabel("year")
plt.ylabel("own")

#adding the legend
plt.show()
```

```python
#line chart with year against own
plt.plot(data["year"])
plt.plot(data["own"])

#adding title to the plot
plt.title("Line Chart ")

#setting the x and y lables
plt.xlabel("year")
plt.ylabel("own")

#adding the legend
plt.show()
```

```python
#bar chart or Bar plot
plt.bar(data["year"], data["own"])

#adding title to the plot
plt.title("Bar Chart")

#setting the x and y lables
plt.xlabel("year")
plt.ylabel("own")

#adding the legend
plt.show()
```

```python
#histogram
plt.hist(data["income"])

#adding title to the plot
plt.title("Histogram")

#adding the legend
plt.show()
```
