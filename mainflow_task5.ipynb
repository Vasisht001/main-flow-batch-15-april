```python
#5. Are there specific age groups that are more prone to heart disease?
pd.crosstab(df.age,df.target)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='target', multiple='stack')
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#at the age of around 57 years, females have a high chance of getting a heart disease and
#males have the highest prabability of having a heart diseasse is at a age of 52. 
```

```python
#6. What is the relationship between maximum heart rate achieved (thalach) and the presence of heart disease?
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='thalach')
plt.title('Maximum Heart Rate by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

```python
#People with high heart rate have a greater counts of heart disease.
```

```python
#7.Is there a relationship between resting blood pressure (trestbps) and heart disease?
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='trestbps', hue="target", palette="inferno")
plt.title('Resting Blood Pressure by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
#there is no significant affect of blood pressure in heart disease.
```

```python
#8.Does fasting blood sugar (fbs) > 120 mg/dl correlate with higher rates of heart disease?
pd.crosstab(df.fbs,df.target)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='fbs', y='target',hue="fbs" ,palette="coolwarm")
plt.title('Fasting Blood Sugar and Heart Disease')
plt.xlabel('Fasting Blood Sugar > 120 mg/dl (0 = No, 1 = Yes)')
plt.ylabel('Proportion with Heart Disease')
plt.show()
```

```python
#if fasting blood sugar is high there is high chance of a heart di
```

```python
#9. How do different resting electrocardiographic results (restecg) impact heart disease prevalence?
pd.crosstab(df.restecg, df.target)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='restecg', hue='target', palette="viridis")
plt.title('Resting ECG Results and Heart Disease')
plt.xlabel('Resting ECG Result')
plt.ylabel('Count')
plt.show()
```

```python
#restecg: Resting electrocardiographic results (0 = normal, 1 = having ST-T wave 
#abnormality, 2 = showing probable or definite left ventricular hypertrophy)

#males with ST-T wave abnormality 33% more chances of heart disease
```

```python
#10. How do systolic blood pressure (trestbps) and cholesterol levels (chol) interact in individuals with and without heart disease?
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='trestbps', y='chol', hue='target',  palette="magma")
plt.title('Systolic Blood Pressure and Cholesterol by Heart Disease Status')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Cholesterol')
plt.show()
```

```python
#we observe that patients with higher cholesterol levels tend to have higher resting blood pressure, 
#and this could be more pronounced in patients with heart disease.
```
